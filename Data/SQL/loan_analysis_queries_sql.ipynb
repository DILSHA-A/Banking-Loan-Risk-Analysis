```python
import pandas as pd
import sqlite3

df = pd.read_csv('loan_data_cleaned.csv')

conn = sqlite3.connect(':memory:')
df.to_sql('loans', conn, index=False, if_exists='replace')

print("Rows loaded:", pd.read_sql("SELECT COUNT(*) FROM loans", conn))
```

```python
query1 = """
SELECT AgeGroup,
       COUNT(*) AS total,
       SUM("Default") AS defaults,
       ROUND(100.0*SUM("Default")/COUNT(*),2) AS default_rate
FROM loans
GROUP BY AgeGroup
ORDER BY default_rate DESC;
"""
print("Query 1: Default Rate by Age Group")
print(pd.read_sql(query1, conn))
```

```python
query2 = """
SELECT IncomeGroup,
       COUNT(*) AS total,
       ROUND(100.0*SUM("Default")/COUNT(*),2) AS default_rate
FROM loans
GROUP BY IncomeGroup
ORDER BY default_rate DESC;
"""
print("Query 2: Default Rate by Income Group")
print(pd.read_sql(query2, conn))
```

```python
query3 = """
SELECT CreditBand,
       COUNT(*) AS total,
       ROUND(100.0*SUM("Default")/COUNT(*),2) AS default_rate
FROM loans
GROUP BY CreditBand
ORDER BY default_rate DESC;
"""
print("Query 3: Default Rate by Credit Score Band")
print(pd.read_sql(query3, conn))
```

```python
query7 = """
SELECT MaritalStatus,
       COUNT(*) AS total,
       ROUND(100.0*SUM("Default")/COUNT(*),2) AS default_rate
FROM loans
GROUP BY MaritalStatus
ORDER BY default_rate DESC;
"""
print("Query 7: Default Rate by Marital Status")
print(pd.read_sql(query7, conn))
```

```python
query8 = """
SELECT HasCoSigner, HasMortgage,
       COUNT(*) AS total,
       ROUND(100.0*SUM("Default")/COUNT(*),2) AS default_rate
FROM loans
GROUP BY HasCoSigner, HasMortgage
ORDER BY default_rate DESC;
"""
print("Query 8: Default Rate by Co-Signer and Mortgage")
print(pd.read_sql(query8, conn))
```

```python
all_queries = f"""-- Query 1: Default Rate by Age Group
{query1}

-- Query 2: Default Rate by Income Group
{query2}

-- Query 3: Default Rate by Credit Score Band
{query3}

-- Query 4: Default Rate by Employment Type
{query4}

-- Query 5: Default Rate by Loan Purpose
{query5}

-- Query 6: Default Rate by Education
{query6}

-- Query 7: Default Rate by Marital Status
{query7}

-- Query 8: Default Rate by Co-Signer and Mortgage
{query8}
"""

with open('loan_analysis_queries.sql', 'w') as f:
    f.write(all_queries)

from google.colab import files
files.download('loan_analysis_queries.sql')
```
